# Graphique de tendance avec une moyenne mobile (Moving Average)

```python
df_thn_pd['Jan_MA'] = df_thn_pd['Jan'].rolling(window=12).mean()  # Moyenne mobile sur 12 mois

plt.figure(figsize=(10, 6))
plt.plot(df_thn_pd['Jan'], label='Température réelle', color='blue')
plt.plot(df_thn_pd['Jan_MA'], label='Moyenne mobile sur 12 mois', color='orange', linestyle='--')
plt.title('Température de janvier avec moyenne mobile')
plt.xlabel('Année')
plt.ylabel('Température')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
# Convertir les colonnes en valeurs numériques, en remplaçant les erreurs de conversion par NaN
df_thn_pd['Jan'] = pd.to_numeric(df_thn_pd['Jan'], errors='coerce')

# Vérifier si des valeurs manquantes existent
print(df_thn_pd['Jan'].isnull().sum())

# Si des valeurs manquantes existent, vous pouvez les traiter (par exemple, en les remplissant par la moyenne)
df_thn_pd['Jan'].fillna(df_thn_pd['Jan'].mean(), inplace=True)

# Vérifiez les valeurs après nettoyage
print(df_thn_pd['Jan'].head())
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Exemple de prévision pour les températures de janvier (THN)
series = df_thn_pd['Jan']  # Série temporelle pour janvier

# Appliquer le modèle ARIMA
model = ARIMA(series, order=(5,1,0))  # Modèle ARIMA
model_fit = model.fit()

# Faire des prévisions
forecast = model_fit.forecast(steps=12)  # Prédire pour les 12 mois suivants
print("Prévisions des températures de janvier:", forecast)
```
